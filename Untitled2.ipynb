```python
import math
import torch
import torch.nn as nn

class InputEmbeddings(nn.Module):
  def __init__(self,d_model:int,vocab_size:int):
    super().__init__()
    self.d_model = d_model
    self.vocab_size = vocab_size
    self.embedding = nn.Embedding(vocab_size,d_model)
  def forward(self,x):
    return self.embedding(x)*math.sqrt(self.d_model)

class PositionalEncoding(nn.Module):
  def __init__(self, d_model:int, seq_len:int, dropout:float) -> None:
    super().__init__()
    self.d_model = d_model
    self.seq_len = seq_len
    self.dropout = nn.Dropout(dropout)

    pe = torch.zeros(seq_len, d_model)

    position = torch.arange(0, seq_len, dtype=torch.float).unsqueeze(1)
    div_term = torch.exp(torch.arange(0, d_model, 2).float() * (-math.log(10000.0) / d_model))

    pe[:, 0::2] = torch.sin(position * div_term)
    pe[:, 1::2] = torch.cos(position * div_term)

    pe = pe.unsqueeze(0) # (1, Seq_len, d_model)
    self.register_buffer("pe", pe)

  def forward(self, x):
    # requires_grad_ hata de, register_buffer isse handle kar raha hai
    x = x + self.pe[:, :x.shape[1], :].detach()
    return self.dropout(x)


class LayerNormalization(nn.Module):
  # features (d_model) pass karna zaroori hai
  def __init__(self, features: int, eps:float=10**-6) -> None:
    super().__init__()
    self.eps = eps
    # 1 ki jagah 'features' size ke parameters lo
    self.alpha = nn.Parameter(torch.ones(features))
    self.bias = nn.Parameter(torch.zeros(features))

  def forward(self, x):
    mean = x.mean(-1, keepdim=True)
    std = x.std(-1, keepdim=True)
    return self.alpha * (x - mean) / (std + self.eps) + self.bias


class FeedForwardBlock(nn.Module):

  def __init__(self, d_model:int, d_ff:int, dropout:float) -> None:
    super().__init__()
    self.linear_1 = nn.Linear(d_model, d_ff) #W1 AND B1
    self.dropout = nn.Dropout(dropout)
    self.linear_2 = nn.Linear(d_ff, d_model)

  def forward(self,x):
    #(Batch,Seq_Len,d_model) -->(Batch,Seq_len,dff) --> (Batch,Seq_len,d_model)
    return (self.linear_2(self.dropout(torch.relu(self.linear_1(x)))))
class MultiHeadAttentionBlock(nn.Module):
  def __init__(self,d_model:int,h:int,dropout:float) -> None:
    super().__init__()
    self.d_model=d_model
    self.h=h
    assert d_model%h==0, "d_model must be divisible by h"

    self.d_k=d_model//h
    self.w_q=nn.Linear(d_model,d_model) #Wq
    self.w_k=nn.Linear(d_model,d_model) #Wk
    self.w_v=nn.Linear(d_model,d_model) #Wv

    self.w_0=nn.Linear(d_model,d_model) #Wo
    self.dropout=nn.Dropout(dropout)

  @staticmethod
  def attention(query,key,value,mask,dropout:nn.Dropout):
    d_k=query.shape[-1]
    # (Batch,h,Seq_len,d_k) -->(Batch,h,Seq_len,Seq-len)
    attention_scores=(query@key.transpose(-2,-1))/math.sqrt(d_k)

    if mask is not None:
      attention_scores=attention_scores.masked_fill(mask==0,-1e9)

    attention_scores=attention_scores.softmax(dim=-1) #(Batch,h,seq_len,seq_len)
    if dropout is not None:
      attention_scores=dropout(attention_scores)
    return (attention_scores@value),attention_scores

  def forward(self,q,k,v,mask):
    query=self.w_q(q) #(Batch,Seq_len,d_model) ---> (Batch,Seq_len,d_model)
    key=self.w_k(k) #(Batch,Seq_len,d_model) ---> (Batch,Seq_len,d_model)
    value=self.w_v(v) #(Batch,Seq_len,d_model) ---> (Batch,Seq_len,d_model)

    #(Batch,Seq_len,d_model) --> (Batch,Seq_len,h,d_k) -->(Batch,h,Seq_len,d_k)
    query=query.view(query.shape[0],query.shape[1],self.h,self.d_k).transpose(1,2)
    key=key.view(key.shape[0],key.shape[1],self.h,self.d_k).transpose(1,2)
    value=value.view(value.shape[0],value.shape[1],self.h,self.d_k).transpose(1,2)

    x, self.attention_scores = MultiHeadAttentionBlock.attention(query, key, value, mask, self.dropout)

    #(Batch,h,Seq_len,d_k) -->(Batch,Seq_len,h,dk) --> (Batch,Seq_len,d_model)
    x=x.transpose(1,2).contiguous().view(x.shape[0],-1,self.h*self.d_k)

    return self.w_0(x)


class ResidualConnection(nn.Module):
  # features (d_model) yahan bhi pass karna zaroori hai
  def __init__(self, features: int, dropout:float) -> None:
    super().__init__()
    self.dropout = nn.Dropout(dropout)
    # Ab ye tere updated LayerNormalization class ke saath match karega
    self.norm = LayerNormalization(features)

  def forward(self, x, sublayer):
    # 'Add & Norm' logic (Pre-norm approach) bilkul sahi hai
    return x + self.dropout(sublayer(self.norm(x)))

class EncoderBlock(nn.Module):
  # features (d_model) yahan bhi pass karna padega
  def __init__(self, features: int, self_attention_block: MultiHeadAttentionBlock, feed_forward_block: FeedForwardBlock, dropout: float) -> None:
    super().__init__()
    self.self_attention_block = self_attention_block
    self.feed_forward_block = feed_forward_block
    # Yahan features (d_model) pass karna zaroori hai
    self.residual_connections = nn.ModuleList([ResidualConnection(features, dropout) for _ in range(2)])

  def forward(self, x, src_mask):
    # Self-Attention sublayer
    x = self.residual_connections[0](x, lambda x: self.self_attention_block(x, x, x, src_mask))
    # Feed-Forward sublayer
    x = self.residual_connections[1](x, self.feed_forward_block)
    return x

class Encoder(nn.Module):
  # features (d_model) ko yahan bhi handle karna hoga
  def __init__(self, features: int, layers: nn.ModuleList) -> None:
    super().__init__()
    self.layers = layers
    # Yahan features (512 ya d_model) pass karo
    self.norm = LayerNormalization(features)

  def forward(self, x, mask):
    for layer in self.layers:
      x = layer(x, mask)
    return self.norm(x)

class DecoderBlock(nn.Module):
  # features (d_model) pass karna zaroori hai
  def __init__(self, features: int, self_attention_block: MultiHeadAttentionBlock, cross_attention_block: MultiHeadAttentionBlock, feed_forward_block: FeedForwardBlock, dropout: float) -> None:
    super().__init__()
    self.self_attention_block = self_attention_block
    self.cross_attention_block = cross_attention_block
    self.feed_forward_block = feed_forward_block
    # Yahan features (d_model) pass karo
    self.residual_connections = nn.ModuleList([ResidualConnection(features, dropout) for _ in range(3)])

  def forward(self, x, encoder_output, src_mask, tgt_mask):
    # 1. Self-Attention (Target mask use hota hai taaki future words na dikhen)
    x = self.residual_connections[0](x, lambda x: self.self_attention_block(x, x, x, tgt_mask))
    # 2. Cross-Attention (Encoder output se context leta hai)
    x = self.residual_connections[1](x, lambda x: self.cross_attention_block(x, encoder_output, encoder_output, src_mask))
    # 3. Feed-Forward
    x = self.residual_connections[2](x, self.feed_forward_block)
    return x


class Decoder(nn.Module):
  # features (d_model) yahan bhi pass karna hoga
  def __init__(self, features: int, layers: nn.ModuleList) -> None:
    super().__init__()
    self.layers = layers
    # Yahan d_model (features) pass karo
    self.norm = LayerNormalization(features)

  def forward(self, x, encoder_output, src_mask, tgt_mask):
    for layer in self.layers:
      # Sequential processing of blocks
      x = layer(x, encoder_output, src_mask, tgt_mask)
    return self.norm(x)

class ProjectionLayer(nn.Module):
  def __init__(self, d_model: int, vocab_size: int) -> None:
    super().__init__()
    self.proj = nn.Linear(d_model, vocab_size)

  def forward(self, x):
    # RETURN RAW LOGITS ONLY!
    # nn.CrossEntropyLoss will handle the math automatically.
    return self.proj(x)

class Transformer(nn.Module):
    def __init__(self, encoder: Encoder, decoder: Decoder, src_embed: InputEmbeddings, tgt_embed: InputEmbeddings, src_pos: PositionalEncoding, tgt_pos: PositionalEncoding, projection_layer: ProjectionLayer) -> None:
        super().__init__()
        self.encoder = encoder
        self.decoder = decoder
        self.src_embed = src_embed
        self.tgt_embed = tgt_embed # Fix: tgt se tgt_embed kiya
        self.src_pos = src_pos     # Fix: Ye assign karna bacha tha
        self.tgt_pos = tgt_pos     # Fix: Ye bhi
        self.projection_layer = projection_layer

    def encode(self, src, src_mask):
        src = self.src_embed(src)
        src = self.src_pos(src)
        return self.encoder(src, src_mask)

    def decode(self, encoder_output, src_mask, tgt, tgt_mask):
        tgt = self.tgt_embed(tgt)
        tgt = self.tgt_pos(tgt)
        return self.decoder(tgt, encoder_output, src_mask, tgt_mask)

    def project(self, x):
        return self.projection_layer(x)


def build_transformer(src_vocab_size, tgt_vocab_size, src_seq_len, tgt_seq_len, d_model=512, N=6, h=8, dropout=0.3, d_ff=2048):
    src_embed = InputEmbeddings(d_model, src_vocab_size)
    tgt_embed = InputEmbeddings(d_model, tgt_vocab_size)
    src_pos = PositionalEncoding(d_model, src_seq_len, dropout)
    tgt_pos = PositionalEncoding(d_model, tgt_seq_len, dropout)

    encoder_blocks = []
    for _ in range(N):
        encoder_self_attn = MultiHeadAttentionBlock(d_model, h, dropout)
        ff = FeedForwardBlock(d_model, d_ff, dropout)
        # Fix: Yahan d_model pass karna zaroori hai
        encoder_blocks.append(EncoderBlock(d_model, encoder_self_attn, ff, dropout))

    decoder_blocks = []
    for _ in range(N):
        decoder_self_attn = MultiHeadAttentionBlock(d_model, h, dropout)
        decoder_cross_attn = MultiHeadAttentionBlock(d_model, h, dropout)
        ff = FeedForwardBlock(d_model, d_ff, dropout)
        # Fix: Yahan bhi d_model pass karna zaroori hai
        decoder_blocks.append(DecoderBlock(d_model, decoder_self_attn, decoder_cross_attn, ff, dropout))

    # Fix: Encoder aur Decoder ko bhi d_model chahiye final normalization ke liye
    encoder = Encoder(d_model, nn.ModuleList(encoder_blocks))
    decoder = Decoder(d_model, nn.ModuleList(decoder_blocks))

    projection_layer = ProjectionLayer(d_model, tgt_vocab_size)

    model = Transformer(encoder, decoder, src_embed, tgt_embed, src_pos, tgt_pos, projection_layer)

    # Xavier initialization bilkul sahi hai (NaN issues rokne ke liye)
    for p in model.parameters():
        if p.dim() > 1:
            nn.init.xavier_uniform_(p)

    return model
```

```python
import torch
from torch.utils.data import Dataset

class CodeSummaryDataset(Dataset):
    def __init__(self, ds, tokenizer_src, tokenizer_tgt, src_len, tgt_len):
        super().__init__()
        self.ds = ds
        self.tokenizer_src = tokenizer_src
        self.tokenizer_tgt = tokenizer_tgt
        self.src_len = src_len
        self.tgt_len = tgt_len

        # Token IDs nikalna (item() use karke int mein convert kiya)
        self.sos_token = torch.tensor([tokenizer_tgt.token_to_id("[SOS]")], dtype=torch.int64)
        self.eos_token = torch.tensor([tokenizer_tgt.token_to_id("[EOS]")], dtype=torch.int64)
        self.pad_token = torch.tensor([tokenizer_tgt.token_to_id("[PAD]")], dtype=torch.int64)

    def __len__(self):
        return len(self.ds)

    def causal_mask(self, size):
        # Upper triangular matrix for look-ahead mask
        mask = torch.triu(torch.ones(1, size, size), diagonal=1).type(torch.int)
        return mask == 0

    def __getitem__(self, idx):
        item = self.ds[idx]
        src_text = item['ast_sequence']
        tgt_text = item['docstring']

        # Tokens mein convert karna
        enc_input_tokens = self.tokenizer_src.encode(src_text).ids
        dec_input_tokens = self.tokenizer_tgt.encode(tgt_text).ids

        # 1. Truncate karna agar tokens limit se zyada hain
        enc_input_tokens = enc_input_tokens[:(self.src_len - 2)] # SOS aur EOS ke liye jagah
        dec_input_tokens = dec_input_tokens[:(self.tgt_len - 1)] # SOS ke liye jagah (Label mein EOS aayega)

        # 2. Padding calculate karna
        enc_num_padding_tokens = self.src_len - len(enc_input_tokens) - 2
        dec_num_padding_tokens = self.tgt_len - len(dec_input_tokens) - 1

        # 3. Padding Tensors banana (Ye sabse safe tarika hai)
        enc_padding = torch.full((enc_num_padding_tokens,), self.pad_token.item(), dtype=torch.int64)
        dec_padding = torch.full((dec_num_padding_tokens,), self.pad_token.item(), dtype=torch.int64)

        # 4. Encoder Input: [SOS] + tokens + [EOS] + [PAD]...
        encoder_input = torch.cat([
            self.sos_token,
            torch.tensor(enc_input_tokens, dtype=torch.int64),
            self.eos_token,
            enc_padding
        ], dim=0)

        # 5. Decoder Input: [SOS] + tokens + [PAD]... (Predicts next word)
        decoder_input = torch.cat([
            self.sos_token,
            torch.tensor(dec_input_tokens, dtype=torch.int64),
            dec_padding
        ], dim=0)

        # 6. Label: tokens + [EOS] + [PAD]... (Humein ye output chahiye)
        label = torch.cat([
            torch.tensor(dec_input_tokens, dtype=torch.int64),
            self.eos_token,
            dec_padding
        ], dim=0)

        return {
            "encoder_input": encoder_input, # (src_len)
            "decoder_input": decoder_input, # (tgt_len)
            "encoder_mask": (encoder_input != self.pad_token.item()).unsqueeze(0).unsqueeze(0).int(), # (1, 1, src_len)
            "decoder_mask": (decoder_input != self.pad_token.item()).unsqueeze(0).unsqueeze(0).int() & self.causal_mask(decoder_input.size(0)),
            "label": label, # (tgt_len)
            "src_text": src_text,
            "tgt_text": tgt_text
        }
```

```python
import torch
import torch.nn as nn
import time
import math
import os
from tqdm import tqdm

# --- 1. SETUP ---
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Optimizer aur Loss Function
pad_idx = tokenizer_tgt.token_to_id('[PAD]')
loss_fn = nn.CrossEntropyLoss(ignore_index=pad_idx, label_smoothing=0.1).to(device)

# Checkpoint path
drive_folder = "/content/drive/MyDrive/ai_summarization_project/checkpoints_aware/"
checkpoint_path = os.path.join(drive_folder, "ast_aware_transformer.pt")

# Resuming Logic
start_epoch = 0
best_val_loss = float('inf')

if os.path.exists(checkpoint_path):
    print(f"🔄 Found checkpoint! Resuming from your last hard work...")
    checkpoint = torch.load(checkpoint_path, map_location=device)
    model.load_state_dict(checkpoint['model_state_dict'])
    optimizer.load_state_dict(checkpoint['optimizer_state_dict'])
    start_epoch = checkpoint['epoch'] + 1
    best_val_loss = checkpoint['loss']
    print(f"✅ Resuming from Epoch {start_epoch + 1}")
else:
    print("⚠️ No checkpoint found, starting from scratch (Epoch 1).")

# --- 2. THE MAIN LOOP ---
EPOCHS = 20

print("\n🚀 STARTING VALUE-AWARE TRAINING...")
print(f"Model Training on: {device}\n")

for epoch in range(start_epoch, EPOCHS):
    # --- TRAINING PHASE ---
    model.train()
    total_train_loss = 0
    train_iterator = tqdm(train_dataloader, desc=f"Epoch {epoch+1}/{EPOCHS} [Train]")

    for batch in train_iterator:
        encoder_input = batch['encoder_input'].to(device)
        decoder_input = batch['decoder_input'].to(device)
        encoder_mask = batch['encoder_mask'].to(device)
        decoder_mask = batch['decoder_mask'].to(device)
        label = batch['label'].to(device)

        # Forward
        encoder_output = model.encode(encoder_input, encoder_mask)
        decoder_output = model.decode(encoder_output, encoder_mask, decoder_input, decoder_mask)
        projection = model.project(decoder_output)

        # Loss
        output_dim = projection.shape[-1]
        loss = loss_fn(projection.reshape(-1, output_dim), label.reshape(-1))

        # Backprop
        optimizer.zero_grad(set_to_none=True)
        loss.backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
        optimizer.step()

        total_train_loss += loss.item()
        train_iterator.set_postfix({"loss": f"{loss.item():.4f}"})

    avg_train_loss = total_train_loss / len(train_dataloader)

    # --- VALIDATION PHASE ---
    model.eval()
    total_val_loss = 0
    val_iterator = tqdm(val_dataloader, desc=f"Epoch {epoch+1}/{EPOCHS} [Val]")

    with torch.no_grad():
        for batch in val_iterator:
            encoder_input = batch['encoder_input'].to(device)
            decoder_input = batch['decoder_input'].to(device)
            encoder_mask = batch['encoder_mask'].to(device)
            decoder_mask = batch['decoder_mask'].to(device)
            label = batch['label'].to(device)

            encoder_output = model.encode(encoder_input, encoder_mask)
            decoder_output = model.decode(encoder_output, encoder_mask, decoder_input, decoder_mask)
            projection = model.project(decoder_output)

            output_dim = projection.shape[-1]
            loss = loss_fn(projection.reshape(-1, output_dim), label.reshape(-1))
            total_val_loss += loss.item()

    avg_val_loss = total_val_loss / len(val_dataloader)

    # --- STATS & SAVE ---
    print(f"\n✅ Epoch {epoch+1} Finished")
    print(f"   Train Loss: {avg_train_loss:.4f} | PPL: {math.exp(min(avg_train_loss, 20)):.2f}")
    print(f"   Val. Loss:  {avg_val_loss:.4f}  | PPL: {math.exp(min(avg_val_loss, 20)):.2f}")

    # Best model save karna
    if avg_val_loss < best_val_loss:
        best_val_loss = avg_val_loss
        torch.save({
            'epoch': epoch,
            'model_state_dict': model.state_dict(),
            'optimizer_state_dict': optimizer.state_dict(),
            'loss': avg_val_loss,
        }, checkpoint_path)
        print(f"💾 Best Checkpoint Updated: {checkpoint_path}")

    # Har epoch ke baad results dekhne ke liye (Tera testing logic yahan aa sakta hai)
    print("-" * 50)
```

```python
import torch
import torch.nn.functional as F
import os
import ast

# --- 0. HELPER: VALUE-AWARE AST CONVERTER ---
def python_to_ast_string(code):
    try:
        tree = ast.parse(code)
        def traverse(node):
            # 1. Get the structural token
            result = [type(node).__name__]

            # 2. Extract the EXACT names and values
            if hasattr(node, 'name'): result.append(str(node.name))
            elif hasattr(node, 'id'): result.append(str(node.id))
            elif hasattr(node, 'arg'): result.append(str(node.arg))
            elif hasattr(node, 'value') and not isinstance(node.value, ast.AST): result.append(str(node.value))

            # 3. Continue down the tree
            for child in ast.iter_child_nodes(node):
                result.extend(traverse(child))
            return result

        return " ".join(traverse(tree))
    except Exception as e:
        return "Module"

# --- 1. UPGRADED BEAM SEARCH (NO LOOPS, MORE LOGIC) ---
def check_model_prediction(model, source_text, tokenizer_src, tokenizer_tgt, device,
                           beam_width=5, temperature=0.7, repetition_penalty=1.5):
    model.eval()
    with torch.no_grad():
        source_ids = tokenizer_src.encode(source_text).ids
        source_tensor = torch.tensor([source_ids], dtype=torch.long).to(device)

        encoder_mask = (source_tensor != tokenizer_src.token_to_id("[PAD]")).unsqueeze(0).unsqueeze(0).int()
        encoder_output = model.encode(source_tensor, encoder_mask)

        sos_idx = tokenizer_tgt.token_to_id("[SOS]")
        eos_idx = tokenizer_tgt.token_to_id("[EOS]")
        pad_idx = tokenizer_tgt.token_to_id("[PAD]")

        beams = [([sos_idx], 0.0)]
        completed_beams = []

        for _ in range(64): # Max generation length
            new_beams = []
            for seq, score in beams:
                if seq[-1] == eos_idx:
                    completed_beams.append((seq, score))
                    continue

                decoder_input = torch.tensor([seq], dtype=torch.long).to(device)
                decoder_mask = (decoder_input != pad_idx).unsqueeze(0).unsqueeze(0).int()

                out = model.decode(encoder_output, encoder_mask, decoder_input, decoder_mask)
                logits = model.project(out[:, -1]) # Shape: [1, vocab_size]

                # 🔥 UPGRADE 1: TEMPERATURE SCALING
                # Softens the distribution so the model isn't overly stubborn about one bad word
                logits = logits / temperature

                # 🔥 UPGRADE 2: HISTORY REPETITION PENALTY
                # Punishes words it has ALREADY used in this sentence to break loops
                for token_id in set(seq):
                    if logits[0, token_id] > 0:
                        logits[0, token_id] /= repetition_penalty
                    else:
                        logits[0, token_id] *= repetition_penalty

                # 🔥 UPGRADE 3: IMMEDIATE REPETITION BLOCKER (Tamed down from -50 to -5)
                if len(seq) > 1:
                    logits[0, seq[-1]] -= 5.0

                log_probs = F.log_softmax(logits, dim=-1)[0]
                top_probs, top_indices = torch.topk(log_probs, beam_width)

                for i in range(beam_width):
                    new_seq = seq + [top_indices[i].item()]
                    new_score = score + top_probs[i].item()
                    new_beams.append((new_seq, new_score))

            if not new_beams:
                break

            # Apply Length Penalty: Score / (length^0.7) to favor slightly longer, descriptive sentences
            new_beams = sorted(new_beams, key=lambda x: x[1] / (len(x[0]) ** 0.7), reverse=True)[:beam_width]
            beams = new_beams

        all_beams = completed_beams + beams
        if len(all_beams) == 0:
            return ""

        all_beams = sorted(all_beams, key=lambda x: x[1] / (len(x[0]) ** 0.7), reverse=True)
        best_seq = all_beams[0][0]

        decoded_text = tokenizer_tgt.decode(best_seq, skip_special_tokens=True)
        return decoded_text

# --- 2. SETUP & LOAD WEIGHTS ---
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
checkpoint_path = "/content/drive/MyDrive/ai_summarization_project/checkpoints_aware/ast_aware_transformer.pt"

if os.path.exists(checkpoint_path):
    print(f"🔄 Loading Weights from: {checkpoint_path}")
    checkpoint = torch.load(checkpoint_path, map_location=device)
    model.load_state_dict(checkpoint['model_state_dict'])
    model.to(device)
    print(f"✅ Model loaded successfully from Epoch {checkpoint.get('epoch', 'Unknown')}!")
else:
    print("❌ Error: Checkpoint file not found. Check the path!")

# --- 3. RUN DEMO (MULTI-SAMPLE) ---
print("\n" + "="*60)
print("🚀 SMART COMPILER: AST TO DOCSTRING EVALUATION (UPGRADED DECODING)")
print("="*60)

sample_codes = [
    # 1. Math / Arithmetic
    "def add(a, b): return a + b",
    "def square(x): return x * x",

    # 2. Logic / Comparison
    "def check_even(n): return n % 2 == 0",
    "def is_admin(user): return user == 'admin'",

    # 3. Operations / IO
    "def greet(name): print('Hello ' + name)",

    # 4. Object Retrieval
    "def get_name(self): return self.name"
]

for i, code_string in enumerate(sample_codes):
    print(f"\n[{i+1}] Evaluating Code: {code_string}")

    ast_input_string = python_to_ast_string(code_string)
    ast_preview = " ".join(ast_input_string.split()[:8])
    print(f"    AST Fed to Model: {ast_preview} ...")

    # Call with the new temperature and penalty parameters
    final_output = check_model_prediction(
        model,
        ast_input_string,
        tokenizer_src,
        tokenizer_tgt,
        device,
        beam_width=5,
        temperature=0.7,        # <-- Magic parameter 1
        repetition_penalty=1.5  # <-- Magic parameter 2
    )

    print(f"    🤖 AI SUMMARY: {final_output}")
    print("-" * 60)
```

```python
def summarize_my_code(code_snippet, model, tokenizer_src, tokenizer_tgt, device, beam_size=3, max_len=64):
    model.eval()

    # 1. AST string to Tokens
    src_tokens = tokenizer_src.encode(code_snippet).ids
    src_input = torch.tensor([src_tokens], dtype=torch.long).to(device)
    encoder_mask = (src_input != tokenizer_src.token_to_id("[PAD]")).unsqueeze(1).unsqueeze(2).to(device)

    with torch.no_grad():
        encoder_output = model.encode(src_input, encoder_mask)

    sos_id = tokenizer_tgt.token_to_id("[SOS]")
    eos_id = tokenizer_tgt.token_to_id("[EOS]")

    # 2. Beam Search Initialization
    # Each beam: (sequence, score)
    beams = [([sos_id], 0.0)]

    for _ in range(max_len):
        new_beams = []
        for seq, score in beams:
            if seq[-1] == eos_id:
                new_beams.append((seq, score))
                continue

            # Prepare decoder input
            decoder_input = torch.tensor([seq], dtype=torch.long).to(device)
            size = decoder_input.size(1)
            tgt_mask = torch.triu(torch.ones((1, size, size)), diagonal=1).type(torch.uint8)
            tgt_mask = (tgt_mask == 0).unsqueeze(0).to(device)

            with torch.no_grad():
                out = model.decode(encoder_output, encoder_mask, decoder_input, tgt_mask)
                logits = model.project(out[:, -1])

                # Apply Repetition Penalty to logits
                for token_id in set(seq):
                    logits[0, token_id] -= 2.0 # Penalty for already used words

                log_probs = torch.log_softmax(logits, dim=-1)

            # Get top k candidates for this beam
            topk_log_probs, topk_ids = torch.topk(log_probs, beam_size)

            for i in range(beam_size):
                new_beams.append((seq + [topk_ids[0][i].item()], score + topk_log_probs[0][i].item()))

        # Keep only the top 'beam_size' beams based on score
        # We divide score by length (Length Penalty) to favor longer, descriptive sentences
        beams = sorted(new_beams, key=lambda x: x[1] / len(x[0]), reverse=True)[:beam_size]

        # If all top beams ended with EOS, stop
        if all(seq[-1] == eos_id for seq, _ in beams):
            break

    # Return the best sequence
    best_seq = beams[0][0]
    # Remove SOS and EOS for clean output
    final_tokens = [t for t in best_seq if t not in [sos_id, eos_id]]

    summary = tokenizer_tgt.decode(final_tokens)

    # Cleanup trailing "raises" loops manually if they persist
    if "raises" in summary:
        summary = summary.split("raises")[0].strip()

    return summary
```

```python
import ast

def get_ast_string(code):
    try:
        tree = ast.parse(code)
        return " ".join([node.__class__.__name__ for node in ast.walk(tree)])
    except Exception:
        return ""

def check_progress(model, tokenizer_src, tokenizer_tgt, device):
    test_functions = [
        "def add(a, b): return a + b",
        "def check_even(n): return n % 2 == 0",
        "def greet(name): print('Hello ' + name)"
    ]

    print(f"\n{'='*20} EPOCH 5 TEST DRIVE {'='*20}")
    for code in test_functions:
        # 1. Convert raw Python to the AST sequence the model understands
        ast_input = get_ast_string(code)

        # 2. Feed the AST sequence to the model
        prediction = summarize_my_code(ast_input, model, tokenizer_src, tokenizer_tgt, device)

        print(f"\nCODE: {code}")
        # Optional: Print a snippet of the AST to verify it converted correctly
        print(f"AST FED TO MODEL: {ast_input[:40]}...")
        print(f"AI SUMMARY: {prediction}")
    print(f"\n{'='*55}")

# Run the test!
check_progress(model, tokenizer_src, tokenizer_tgt, device)
```

```python
def check_model_prediction(model, source_text, tokenizer_src, tokenizer_tgt, device):
    model.eval()
    with torch.no_grad():
        # Source text ko tokens mein badalna
        source_ids = tokenizer_src.encode(source_text).ids
        source_tensor = torch.tensor([source_ids], dtype=torch.long).to(device)

        # Encoder mask banana
        encoder_mask = (source_tensor != tokenizer_src.token_to_id("[PAD]")).unsqueeze(0).unsqueeze(0).int()

        # Encoder se hidden states nikalna
        encoder_output = model.encode(source_tensor, encoder_mask)

        # Decoder ko start token se shuru karna
        decoder_input = torch.tensor([[tokenizer_tgt.token_to_id("[SOS]")]], dtype=torch.long).to(device)

        # Ek-ek karke next word predict karna (Greedy Search)
        for _ in range(64): # max length
            decoder_mask = (decoder_input != tokenizer_tgt.token_to_id("[PAD]")).unsqueeze(0).unsqueeze(0).int()
            # Causal mask add karna (Hamare custom dataset wala logic)
            # Yahan hum simplified inference kar rahe hain

            out = model.decode(encoder_output, encoder_mask, decoder_input, decoder_mask)
            prob = model.project(out[:, -1])
            next_word = torch.argmax(prob, dim=-1)

            decoder_input = torch.cat([decoder_input, torch.tensor([[next_word]], device=device)], dim=1)

            if next_word == tokenizer_tgt.token_to_id("[EOS]"):
                break

        # IDs ko wapas text mein badalna
        tokens = decoder_input[0].tolist()
        decoded_text = tokenizer_tgt.decode(tokens, skip_special_tokens=True)
        return decoded_text
```

```python
def get_smart_prediction(model, code, tokenizer_src, tokenizer_tgt, device, max_len=10):
    model.eval()
    # 1. Encode Source
    src_tokens = tokenizer_src.encode(code).ids
    src_tensor = torch.tensor([src_tokens]).to(device)
    src_mask = (src_tensor != tokenizer_src.token_to_id("[PAD]")).unsqueeze(0).unsqueeze(0).int()

    encoder_output = model.encode(src_tensor, src_mask)

    # 2. Start Decoding
    decoder_input = torch.tensor([[tokenizer_tgt.token_to_id("[SOS]")]], dtype=torch.long).to(device)
    generated_tokens = []

    for _ in range(max_len):
        decoder_mask = (decoder_input != tokenizer_tgt.token_to_id("[PAD]")).unsqueeze(0).unsqueeze(0).int()
        out = model.decode(encoder_output, src_mask, decoder_input, decoder_mask)
        prob = model.project(out[:, -1])

        # --- REPETITION PENALTY ---
        # Jo word aa gaya, uski probability kam kar do taaki loop na bane
        for token in set(generated_tokens):
            prob[0, token] /= 2.0

        next_word = torch.argmax(prob, dim=-1).item()
        generated_tokens.append(next_word)

        if next_word == tokenizer_tgt.token_to_id("[EOS]"):
            break

        decoder_input = torch.cat([decoder_input, torch.tensor([[next_word]], device=device)], dim=1)

    return tokenizer_tgt.decode(generated_tokens, skip_special_tokens=True)

# Test karo
print(f"✨ NEW SMART RESULT: {get_smart_prediction(model, 'def square(x): return x * x', tokenizer_src, tokenizer_tgt, device)}")
```
